# Exploratory data analysis

Descriptive statistics

```python
train_data_frame.describe().transpose()
```

Output variables investigation

```python
train_data_frame['SalePrice'].describe()
```

```python
print(f"Skewness: {train_data_frame['SalePrice'].skew()}")
print(f"Kurtosis: {train_data_frame['SalePrice'].kurt()}")
```

```python
hist, bins = np.histogram(train_data_frame['SalePrice'].to_numpy(), bins=50,density=True)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width, zorder=1)

# density plot
train_data_frame['SalePrice'].plot.kde(zorder=2, color='C1')
plt.show()
```

Correlation between numerical variables

```python
corr = train_data_frame.corr()
corr.style.background_gradient(cmap='coolwarm')
```
